```python
# 导入必要的库
import pandas as pd
import matplotlib.pyplot as plt
import os

%matplotlib inline
name = 'traffic'
data = pd.read_csv(f'./dataset/traffic/{name}.csv')
data_head = 2500 # 部分数据

# 3. 检查是否有'OT'列
if 'OT' not in data.columns:
    print("\n错误：数据集中没有'OT'列！")
    print("可用列包括：", data.columns.tolist())
else:
    # 4. 绘制'OT'列走势图
    data['date'] = pd.to_datetime(data['date'])
    plt.figure(figsize=(12, 6))
    plt.plot(data['date'],data['OT'], label='OT')
    plt.title(f'{name} Dataset - OT Column Trend')
    plt.xlabel('Time')
    plt.ylabel('OT Value')
    plt.legend()
    plt.grid(True)
    plt.gcf().autofmt_xdate()

    if not os.path.exists('./pic/plots'):
        os.makedirs('./pic/plots')
    plt.savefig(f'./pic/plots/{name}.svg', dpi=100)
    plt.show()
```

```python
import matplotlib.pyplot as plt
import pywt
name = 'ettm1'
data = pd.read_csv(f'./dataset/ETT-small/{name}.csv')
data_head = 1000 # 部分数据
subset = data.head(data_head).copy()
subset['date'] = pd.to_datetime(subset['date'])
# 假设已经进行了小波分解
coeffs_list = pywt.wavedec(subset['OT'], level=3, wavelet='db4')

# 设置画布
plt.figure(figsize=(12, 8))

# 绘制原始信号
plt.subplot(len(coeffs_list) + 1, 1, 1)  # (rows, cols, index)
plt.plot(subset['OT'], 'b', label='Original Signal')
plt.axis('off')
plt.savefig(f'./pic/plots/{name}_origin.svg')

# 绘制各层小波系数
titles = ['Approximation (A3)', 'Detail (D3)', 'Detail (D2)', 'Detail (D1)']
for i, (coeff, title) in enumerate(zip(coeffs_list, titles)):
    plt.figure(figsize=(12, 8))
    plt.subplot(len(coeffs_list) + 1, 1, i + 2)  # 从第2行开始
    plt.plot(coeff, 'r', label=title)
    plt.axis('off')
    plt.savefig(f'./pic/plots/{name}_level_{i}.svg')
plt.tight_layout()  # 防止子图重叠
plt.show()
```
